```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
# Remove rows with missing values
df = df.dropna()
```

```python
# Identify numerical and categorical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = df.select_dtypes(include=['object']).columns.tolist()

# Print the results
print("Numerical Features:")
print(numerical_features)
print("\nCategorical Features:")
print(categorical_features)
```

```python
df = df[(df['Electric Range'] > 0) & (df['Base MSRP'] > 0)]
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
df.info()
```

```python
print("\nNumerical features:", numerical_features)
print("Categorical features:", categorical_features)
```

```python
import plotly.express as px

# Scatter plot: electric range vs. Base MSRP
fig = px.scatter(df, x='Electric Range', y='Base MSRP', color='Make',
                 title='Electric Range vs Base MSRP by Make')
fig.show()
```

```python
# Box plot: Base MSRP by Make
fig = px.box(df, x='Make', y='Base MSRP', color='Make',
             title='Base MSRP Distribution by Make')
fig.show()
```

```python
# Pie chart: Distribution of Electric Vehicle Types
fig = px.pie(df, names='Electric Vehicle Type', title='Distribution of Electric Vehicle Types')
fig.show()
```

```python
# Choropleth plot: Number of electric vehicles by state
df_state = df.groupby('State').size().reset_index(name='count')
fig = px.choropleth(df_state, locations='State', locationmode='USA-states', color='count',
                    scope='usa', title='Electric Vehicle Distribution by State')
fig.show()
```

```python
# Animated Choropleth: Changes in Electric Vehicle Distribution by State and Model Year
df_year_state = df.groupby(['State', 'Model Year']).size().reset_index(name='count')
fig = px.choropleth(df_year_state, locations='State', locationmode='USA-states', color='count',
                    scope='usa', animation_frame='Model Year', title='Changes in Electric Vehicle Distribution by State and Model Year')
fig.show()
```

```python
df_grouped = df.groupby(['Model Year', 'Make']).size().unstack(fill_value=0)

# Check the structure of the grouped dataframe
print(df_grouped.head())
```

```python
import bar_chart_race as bcr

# Create the bar chart race animation
bcr.bar_chart_race(
    df=df_grouped,  # The grouped dataframe where rows are 'Model Year' and columns are 'Make'

    # Output filename, set to None to display directly without saving
    filename="electric_vehicle_race.mp4",

    # Orientation of the bars: horizontal
    orientation="h",

    # Sorting the bars in descending order
    sort="desc",

    # Number of bars to display in each frame
    n_bars=10,

    # Smoothness of the animation (higher = smoother)
    steps_per_period=45,

    # Duration for each frame (in milliseconds)
    period_length=3000,

    # Format of the period (year) label text
    period_fmt='{x:.0f}',
)
```
