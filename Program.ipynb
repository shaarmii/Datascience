```python
movie_rating_counts = ratings_df.groupby('movieId')['rating'].count()
sorted_movies = movie_rating_counts.sort_values(ascending=False)
top_5_movie_ids = sorted_movies.head(5).index
top_5_movie_titles = movies_df[movies_df['movieId'].isin(top_5_movie_ids)]['title']

print("Top 5 popular movies based on number of user ratings:")
for title in top_5_movie_titles:
  print(title)
```

```python

grouped_ratings = ratings_df.groupby('movieId')['rating'].agg(['count'])
merged_df = grouped_ratings.merge(movies_df, on='movieId')
filtered_movies = merged_df[merged_df['count'] > 50]

sorted_movies = filtered_movies.sort_values('count', ascending=False)
sci_fi_movies = sorted_movies[sorted_movies['genres'].str.contains('Sci-Fi')]

if len(sci_fi_movies) >= 3:
    third_most_popular_sci_fi = sci_fi_movies.iloc[2]['title']
    print("The third most popular Sci-Fi movie based on number of user ratings is:", third_most_popular_sci_fi)
else:
    print("There are not enough Sci-Fi movies with over 50 ratings to determine the third most popular.")
```
