1. **Data Description**

This dataset contains information about job candidates and their salaries.
2. **Objective of the Project**

The goal of this project is to explore what factors influence job candidates' salaries, including education, specialization, and exam scores.

```python
date_columns = ['DOJ', 'DOL', 'DOB']
for date_column in date_columns:
    data[date_column] = pd.to_datetime(data[date_column], errors='coerce')
```

```python
data[date_columns].head()
```

```python
# Convert categorical columns to category type
categorical_columns = ['Designation', 'JobCity', 'Gender', '10board', '12board', 'Degree', 'Specialization', 'CollegeState']
for cat_column in categorical_columns:
    data[cat_column] = data[cat_column].astype('category')
```

```python
data[categorical_columns]
```

```python
print("Duplicates after processing:", data.duplicated().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example for Salary
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Salary'])
plt.title('Boxplot of Salary')
plt.show()
```

```python
# Identify outliers in Salary
Q1 = data['Salary'].quantile(0.25)
Q3 = data['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['Salary'] < lower_bound) | (data['Salary'] > upper_bound)]
print("Outliers in Salary:\n", outliers)
```

```python
# Display only relevant columns for the outliers
relevant_columns = ['ID', 'Salary', 'Designation', 'JobCity', 'Gender', 'DOB', '10percentage']
print(outliers[relevant_columns])
```

```python
# Grouping outliers by Designation and calculating mean salary
designation_salary = outliers.groupby('Designation')['Salary'].mean().sort_values(ascending=False)
print(designation_salary)
```

```python
# Filter out NaN values
designation_salary = designation_salary[designation_salary.notna()]
print(designation_salary)
```

```python
plt.figure(figsize=(15, 6))  # Increased figure size
sns.barplot(x=designation_salary.index[:20], y=designation_salary.values[:20])  # Only top 20 for clarity
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.title('Mean Salary by Designation for Outliers', fontsize=16)  # Add title with larger font size
plt.xlabel('Designation', fontsize=14)  # Add x-label with larger font size
plt.ylabel('Mean Salary', fontsize=14)  # Add y-label with larger font size
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='collegeGPA', y='Salary', alpha=0.7)
plt.title('Scatter Plot of Salary vs. College GPA', fontsize=16)
plt.xlabel('College GPA', fontsize=14)
plt.ylabel('Salary (in INR)', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(data['10percentage'], data['Salary'], gridsize=30, cmap='Blues', mincnt=1)
plt.colorbar(label='Count in Hexbin')
plt.title('Hexbin Plot of Salary vs. 10 Percentage', fontsize=16)
plt.xlabel('10 Percentage (Overall Marks)', fontsize=14)
plt.ylabel('Salary (in INR)', fontsize=14)
plt.grid(False)  # Remove gridlines for clarity
plt.tight_layout()
plt.show()
```

```python
# Select relevant numerical columns for pair plot
numerical_cols = ['Salary', 'collegeGPA', '10percentage', 'English', 'Logical']
sns.pairplot(data[numerical_cols], diag_kind='kde', markers='o', plot_kws={'alpha':0.6})
plt.suptitle('Pair Plot of Selected Continuous Variables', y=1.02, fontsize=16)
plt.show()
```

```python
# Filter for relevant job roles
job_roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
salary_test_df = data[data['Designation'].isin(job_roles)]

# Calculate the average salary
average_salary = salary_test_df['Salary'].mean()

print(f"Average Salary for specified job roles: {average_salary:.2f} INR")
print(f"Claimed Salary Range: 2.5-3 Lakhs INR")
```

```python
contingency_table = pd.crosstab(data['Gender'], data['Specialization'])

# Visualization
plt.figure(figsize=(12, 8))  # Adjust figure size for better visibility
contingency_table.plot(kind='bar', stacked=True)

plt.title('Gender Preference by Specialization', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0)

# Move the legend outside the plot
plt.legend(title='Specialization', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(axis='y')
plt.tight_layout()  # Adjust layout to fit elements properly
plt.show()
```
